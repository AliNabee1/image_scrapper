```python
from __future__ import unicode_literals
import urllib.request
from selenium import webdriver
from selenium.webdriver.firefox import options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import requests
import csv
from time import sleep
import random
import time
import os
import pandas as pd
driver = webdriver.Chrome()
```

```python
driver.get('https://www.houzz.com/professionals/general-contractor')
```

```python
#step 1
count=0
imgs_link=[]
while True:    
    picsss=driver.find_elements_by_xpath("//div[@class='hz-pro-profile-info-rating clearfix']/a[@href]")
    
    for image in picsss:
        try:
            imgs_link.append(image.get_attribute('href'))
        except:
            ()
        sleep(1)
    print(imgs_link)
    try:
        
        driver.find_elements_by_xpath("//span/span[contains(text(),'Next Page')]")[0].click()
        sleep(2)
    except:
        break
```

```python
count=0
for img in imgs_link:
    driver.get(img)
    pics=driver.find_elements_by_xpath("//div/a[@href and @data-testid='image-card-link']")
    imgs_linkk=[]
    for imagee in pics:
        try:
            imgs_linkk.append(imagee.get_attribute('href'))
        except:
            ()
        sleep(3)
    print(imgs_linkk)
    for imgg in imgs_linkk:
        driver.get(imgg)
        sleep(3)
        
        try:
            driver.find_element_by_xpath("//div/picture/img[@src]").click()
            sleep(4)
        except:
            ()
        #image=driver.find_element_by_xpath("//div/div[2]/img[@src and @class='view-photo-image-pane__image']")
        while True:
            try:
                sleep(2)
                image=driver.find_element_by_xpath("//div/div[2]/img[@src and @class='view-photo-image-pane__image']")
                src=image.get_attribute('src') 
                count=count+1
                try:
                    img_name="houzz/gerneralcontractor/"+"img_"+str(count)+".jpg"
                    urllib.request.urlretrieve(src,img_name)
                except:
                    ()
                nextt=driver.find_element_by_xpath("//div//button[@aria-label='Next photo']/span")
                driver.execute_script("arguments[0].click();",nextt)
            except:
                break
```
